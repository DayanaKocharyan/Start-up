```python
df['Employee Count'] = df['Employee Count'].astype('Int64')
```

```python
def employee_count(data):
    if pd.isnull(data['Team size all employees']) and pd.notnull(data['Employees count MoM change']):
        data['Team size all employees'] = data['Employee Count'] + int(data['Employees count MoM change'])
    if pd.isnull(data['Team size all employees']) and pd.isnull(data['Employees count MoM change']):
        data['Employees count MoM change'] = 0
        data['Team size all employees'] = data['Employee Count']
    if pd.notnull(data['Team size all employees']):
        data['Employees count MoM change'] = int(data['Team size all employees']) - data['Employee Count']
    return data
```

```python
df = df.apply(employee_count, axis=1)
```

```python
df[['Employee Count', 'Employees count MoM change', 'Team size all employees']].info()
```

```python
def team_size_grown(data):
    if data['Employees count MoM change'] > 0:
        data['team_size_grown']= 'Yes'
    else:
        data['team_size_grown']= 'No'
    return data
```
